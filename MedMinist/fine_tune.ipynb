```python
def test(model, test_loader):
    model.eval()
    test_total = 0
    test_correct = 0
    # Disable gradient calculation for faster inference
    with torch.no_grad():
        for inputs, targets in test_loader:
            inputs, targets = inputs.to(device), targets.to(device)
            targets = targets.squeeze().long()
            # Forward pass
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            # Update correct and total counts
            test_total += targets.size(0)
            test_correct += (predicted == targets).sum().item()
    # Calculate accuracy
    test_accuracy = 100 * test_correct / test_total
    print(f'Test Accuracy: {test_accuracy:.2f}%')
    return test_accuracy
```

```python
#消融训练
import copy
from torch import optim, nn
from tqdm import tqdm
import torch
from MedMinist.Ablation_model import DEN_V3

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
num_epochs = 7
lr = 0.0001
num_classes = 9

# 用于存储测试准确率结果
results = {}

# 渐进式消融的设置函数
def apply_ablation(model, ablation_count):
    """
    动态设置模型的 DEB 模块的 ablation 状态。
    :param model: DEN_V3 模型实例
    :param ablation_count: 要禁用合作模式的层数
    """
    count = 0
    for layer in [model.layer1, model.layer2, model.layer3, model.layer4]:
        for block in layer:
            if count < ablation_count:
                block.deep_explain.ablate = True  # 启用非合作模式
                count += 1
            else:
                block.deep_explain.ablate = False  # 保持合作模式

# 创建基础模型
denv6_base = DEN_V3(num_classes, inchannels=3).to(device)
model = torch.load('path_DEN_V6_93.50.pth', map_location=device)
denv6_base.load_state_dict(model.state_dict())  # 加载预训练权重

# 定义损失函数
criterion = nn.CrossEntropyLoss()

# 记录最佳模型
best_model = None
max_acc = 0

# 开始训练每个 ablation 模型
for ablation_count in range(1, 9):  # 从 ablate1 到 ablate8
    print(f"\n--- Training model with ablation_count={ablation_count} ---")
    # 创建模型副本
    model = copy.deepcopy(denv6_base)
    apply_ablation(model, ablation_count)  # 应用 ablation

    optimizer = optim.SGD(model.parameters(), lr=lr, momentum=0.9)

    # 训练模型
    for epoch in range(num_epochs):
        model.train()
        for inputs, targets in tqdm(train_loader, disable=True):
            optimizer.zero_grad()
            inputs, targets = inputs.to(device), targets.to(device)
            targets = targets.squeeze().long()
            outputs = model(inputs)

            loss = 0
            try:
                if isinstance(outputs, list):
                    for output in outputs:
                        loss += criterion(output, targets)
                else:
                    loss = criterion(outputs, targets)
            except Exception as e:
                print(e)
                continue
            loss.backward()
            optimizer.step()

    # 测试模型性能
    acc = test(model, test_loader)
    print(f"Ablation {ablation_count}: Test accuracy = {acc:.2f}%")
    results[f"ablate{ablation_count}"] = acc

    # 更新最佳模型
    if ablation_count == 8:  # 如果是 ablate8，记录为最佳模型
        best_model = copy.deepcopy(model)
        max_acc = acc

# 输出所有结果
print("\nFinal Results:")
for key, value in results.items():
    print(f"{key}: {value:.2f}%")

# 保存最佳模型（可选）
torch.save(best_model.state_dict(), 'path_best_model_ablate8.pth')
```

```python
from scipy.stats import pearsonr
import torch
import numpy as np

# 存储每个样本的数据
all_similarities = []  # 每张图片的7个相似度值
all_average_drops = []  # 每张图片的 Average Drop 值

def compute_drop_soft(img_tensor, label, resnet34, weight, device):
    """
    根据你给定的计算方法来计算 drop_soft 值。
    img_tensor: shape [1, C, H, W] 的张量（单张图片）
    label: 标签 (int or tensor)，需要与网络输出维度匹配
    resnet34: 模型实例，用于预测
    weight: 显著性强度的权值(0-1之间)
    device: 计算设备
    """

    img_tensor = img_tensor.to(device)
    label = torch.tensor([label], device=device) if isinstance(label, int) else label.to(device)
    
    # 原始预测概率 p_orig
    with torch.no_grad():
        outputs_orig = resnet34(img_tensor)
        p_orig = torch.softmax(outputs_orig, dim=1)[0, label].item()

    # 计算均值图像
    img_mean = img_tensor.mean(dim=(2, 3), keepdim=True)  # [1, C, 1, 1]

    # Soft替换
    img_soft_masked = img_tensor.clone()
    img_soft_masked = img_soft_masked * (1 - weight) + img_mean * weight

    # 计算 p_soft_masked
    with torch.no_grad():
        outputs_soft = resnet34(img_soft_masked)
        p_soft_masked = torch.softmax(outputs_soft, dim=1)[0, label].item()

    drop_soft = max(0, (p_orig - p_soft_masked) / p_orig)

    return drop_soft

def compute_average_drop(model, input_tensor, target, criterion, device):
    """
    计算单个样本的 Average Drop 值
    """
    model.eval()
    input_tensor = input_tensor.to(device)
    target = target.to(device)

    # 原始输出
    with torch.no_grad():
        outputs = model(input_tensor)
        original_pred = outputs[-1]  # 最后的预测结果
        original_score = torch.softmax(original_pred, dim=1)[0, target].item()

    # 删除显著区域后的输出
    # 使用 Grad-CAM 等方法生成显著性图（这里只是示意）
    importance_mask = torch.ones_like(input_tensor) * 0.5  # 伪显著性图，实际应替换为 Grad-CAM 结果
    modified_input = input_tensor * importance_mask

    with torch.no_grad():
        modified_outputs = model(modified_input)
        modified_score = torch.softmax(modified_outputs[-1], dim=1)[0, target].item()

    # 计算 Average Drop
    average_drop = max(0, original_score - modified_score) / original_score
    return average_drop

# 遍历整个训练集
for iter_idx, (inputs, targets) in enumerate(train_loader):
    inputs, targets = inputs.to(device), targets.to(device)

    for img_idx in range(inputs.size(0)):  # 遍历 batch 中的每张图片
        input_tensor = inputs[img_idx].unsqueeze(0)
        target = targets[img_idx].unsqueeze(0)

        # 1. 计算前7层的相似度
        similarities = compute_first_7_block_similarity(model, input_tensor, target, criterion, device)
        all_similarities.append(similarities)

        # 2. 计算 Average Drop
        average_drop = compute_average_drop(model, input_tensor, target, criterion, device)
        all_average_drops.append(average_drop)

# 3. 计算相关性
all_similarities = np.array(all_similarities)  # (num_samples, 7)
all_average_drops = np.array(all_average_drops)  # (num_samples, )

correlations = []
for block_idx in range(7):
    similarity_values = all_similarities[:, block_idx]
    correlation, p_value = pearsonr(similarity_values, all_average_drops)
    correlations.append((correlation, p_value))
    print(f"Block {block_idx + 1} 相似度与 Average Drop 的相关性: {correlation:.4f}, p值: {p_value:.4f}")

# 4. 可视化
import matplotlib.pyplot as plt

blocks = [f"Block {i+1}" for i in range(7)]
correlation_values = [c[0] for c in correlations]

plt.figure(figsize=(10, 6))
plt.bar(blocks, correlation_values, color='skyblue')
plt.title("Correlation between Gradient Similarity and Average Drop")
plt.xlabel("Block")
plt.ylabel("Pearson Correlation Coefficient")
plt.ylim(-1, 1)
plt.grid(alpha=0.5, linestyle='--')
plt.show()
```
